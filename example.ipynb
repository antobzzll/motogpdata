## `api.Season()`

```python
from motogpdata import api

print(f"Last 10 seasons: {api.list_seasons()[:10]}")
print(f"First 10 seasons: {api.list_seasons()[:-11:-1]}")
```

```python
season = api.Season(
    season=2022,
    category='MotoGP',
    verbose=True
)

print(f"Selected season year: {season.selected_season_year}")
print(f"Selected season category: {season.selected_cat_name}")
# print(f"Available events in selected season:\n{season.events_list}")
```
