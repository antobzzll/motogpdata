# MotoGP 2022 -- Francesco Bagnaia vs Fabio Quartararo
Who's the rider with the highest chance of winning the MotoGP 2022 Championship?

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../data/2022.csv")
bagnaia = df.loc[df['rider.full_name'] == "Francesco Bagnaia"]
quartararo = df.loc[df['rider.full_name'] == "Fabio Quartararo"]
```

```python
fig, ax = plt.subplots(ncols=2, figsize=(15, 5))
sns.kdeplot(df, x='position', color='grey', label='All riders', ax=ax[0])
sns.kdeplot(bagnaia, x='position', color='red', label='Francesco Bagnaia', ax=ax[0])
sns.kdeplot(quartararo, x='position', color='blue', label='Fabio Quartararo', ax=ax[0])
ax[0].set_title("Positions")
ax[0].legend()

sns.kdeplot(df, x='points', color='grey', label='All riders', ax=ax[1])
sns.kdeplot(bagnaia, x='points', color='red', label='Francesco Bagnaia', ax=ax[1])
sns.kdeplot(quartararo, x='points', color='blue', label='Fabio Quartararo', ax=ax[1])
ax[1].set_title("Points")
ax[1].legend()

fig.suptitle("Distribution comparison of positions and points")
plt.show()
```

```python
fig, ax = plt.subplots(ncols=2, figsize=(15, 5))
sns.histplot(bagnaia, x='position', color='red', label='Francesco Bagnaia', ax=ax[0], bins=25)
sns.histplot(quartararo, x='position', color='blue', label='Fabio Quartararo', ax=ax[0], bins=25)
ax[0].set_title("Positions")
ax[0].legend()

sns.histplot(bagnaia, x='points', color='red', label='Francesco Bagnaia', ax=ax[1], bins=25, discrete=True)
sns.histplot(quartararo, x='points', color='blue', label='Fabio Quartararo', ax=ax[1], bins=25, discrete=True)
ax[1].set_title("Points")
ax[1].legend()

fig.suptitle("Comparison of positions and points")
plt.show()
```
